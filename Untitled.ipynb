```python
import numpy as np;
import scipy as sp;
import matplotlib.pyplot as plt
from scipy import signal
```

```python
# Primer : Upravljanje DC motorom  G(s)=10/s(s+2)(s+8)
# postaviti polove procesa na p1=-1.4 p2,3=-1+- 2j  , polove estimatora na pe1=-4.2 , pe2,3=-3+-6.4j
# Opservabilna kanonicna forma:
A = np.matrix([[-10.,1.,0.],[-16.,0.,1.],[0.,0.,0.]])
B = np.matrix([[0.],[0.],[10.]])
C = np.matrix([[1.,0.,0.]])
D = [[0.]]
G = signal.lti(A,B,C,D)
t=np.linspace(0,10,1000)
u=(t>0)
tt,y,x= signal.lsim(G,u,t)
plt.plot(tt,x[:,0],'r',tt,x[:,1],'b',tt,x[:,2],'g',tt,y,'y')
plt.grid()
plt.show()
plt.figure()
plt.plot(x[:,0],x[:,1])
plt.grid()
plt.show()
```

```python
P=np.array([-1.4,-1+ 2*1j,-1- 2*1j])
fsf1=signal.place_poles(A,B,P)
K= fsf1.gain_matrix
K
```

```python
P= np.array([-4.2,-3+6.4*1j,-3-6.4*1j])
fsf1=signal.place_poles(A.T,C.T,P)
L=fsf1.gain_matrix
L=L.T
L
```

```python
A1u=np.concatenate((A-B*K,B*K),axis=1)
A2u=np.concatenate((np.matrix([[0,0,0],[0,0,0],[0,0,0]]),A-L*C),axis=1)
Au= np.concatenate((A1u,A2u),axis=0)
Bu= np.concatenate((B,B),axis=0)
Cu= np.matrix([[1,0,0,0,0,0]])
Du= 0
G= signal.lti(Au,Bu,Cu,Du)
t=np.linspace(0,5,1000)
u=(t>0)

tt,y,x= signal.lsim(G,u,t)
plt.plot(tt,x[:,0:3],'b',tt,x[:,1],tt,x[:,3:6],'r')
plt.grid()
plt.show()

plt.figure()
xo=[1,1,1,1,1,1]
tt,y,x= signal.lsim(G,U=None,T=t,X0=xo)
plt.plot(tt,x[:,0:3],'b',tt,x[:,1],tt,x[:,3:6],'r')
plt.grid()
plt.show()
```
