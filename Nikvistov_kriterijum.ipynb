```python
import numpy as np;
import scipy as sp;
import matplotlib.pyplot as plt
from scipy import signal
```

```python
#Zadatak 1: 
#a) Nacrtati Bodeove dijagrame funkcije prenosa W(s)=1/(s+1)^2
#b) Nacrtati Nikvistov dijagram
#c) Ispitati stabilnost funkcije spregnutog prenosa u zavisnosti od pojacanja K
```

```python
W=signal.lti([], [-1, -1], 1)
# a)
w_log=np.logspace(-2,4,num=1000)
w, mag, phase = signal.bode(W,w=w_log)
plt.figure()
plt.grid()
plt.semilogx(w, mag)    # Bode amplitudska karakteristika
plt.figure()
plt.grid()
plt.semilogx(w, phase)  # Bode fazna karakteristika
plt.show()
```

```python
#b) 
ww, H = signal.freqresp(W)
plt.figure()
plt.grid()
plt.plot(H.real, H.imag, "b")
plt.plot(H.real, -H.imag, "r")
plt.show()
```

```python
#c) sistem je stabilan za svako K
```

```python
#Zadatak 6: 
#Ponoviti za W(s)=85(s+1)(s^2+2s+43.25)/s^2(s^2+2s+82)(s^2+2s+101)
```

```python
brojilac=sp.convolve(sp.convolve([1,2,43.25],[1,1]),[85])
print(brojilac)
imenilac=sp.convolve(sp.convolve([1,0,0],[1,2,82]),[1,2,101])
print(imenilac)
```

```python
W=signal.lti(brojilac,imenilac)
w_log=np.logspace(-0.3,3,num=1000)
ww, H = signal.freqresp(W,w=w_log)
plt.figure()
plt.grid()
plt.plot(H.real, H.imag, "b")
plt.plot(H.real, -H.imag, "r")
plt.show()
```

```python
# Fazna margina je velika, amplitudska ukazuje na losu marginu stabilnosti!
```
