```python
from scipy import signal
```

```python
import scipy as sp
```

```python
G = signal.lti([1], [1, 5])
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
t = sp.linspace(0, 50, 1000)
```

```python
t1, u = G.step()
```

```python
plt.plot(t1, u)
```

```python
x=sp.cos(2*sp.pi*3*t)
```

```python
plt.plot(t, x)
```

```python
tout, y, y1 = signal.lsim(G, x, t)
```

```python
len(y)
```

```python
plt.plot(tout, y, t, x)
```

```python
plt.plot(tout, y)
```

```python
w,Gjw = G.freqresp(n=1000)
```

```python
Gjw
```

```python
plt.plot(w, abs(Gjw))
```

```python
plt.plot(w/2/sp.pi, abs(Gjw))
```

```python
x2 = sp.cos(2*sp.pi*0.1*t)
```

```python
tout2, y22, y23 = signal.lsim(G, x2, t)
```

```python
plt.plot(tout2, y22, t, x2)
```
