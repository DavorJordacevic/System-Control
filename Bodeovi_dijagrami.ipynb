```python
import numpy as np;
import scipy as sp;
import matplotlib.pyplot as plt
from scipy import signal
```

```python
#Zadatak 1: 
#a) Nacrtati Bodeove dijagrame funkcije prenosa (prvo na papiru) W(s)=2000/s(s+200)
#b) Za K=1 oderediti presecnu ucestanost pojacanja, presecnu ucestanost faze i faznu i amplitudsku marginu
#c) Ispitati stabilnost sistema sa zatvorenom povratnom spregom u zavisnosti od K
#d) Odrediti K tako da je fazna margina 30 stepeni
#e) Nacrtati Bodeove dijagrame funkcije prenosa KW(s) za K izracunato u d)
```

```python
W=signal.lti([], [0, -200], 2000)
```

```python
# a)
w_log=np.logspace(-2,4,num=1000)
w, mag, phase = signal.bode(W,w=w_log)
plt.figure()
plt.grid()
plt.semilogx(w, mag)    # Bode amplitudska karakteristika
plt.figure()
plt.grid()
plt.semilogx(w, phase)  # Bode fazna karakteristika
plt.show()
```

```python
# b) presecna ucestanost pojacanja
w1=w[mag[:]>=0][-1]
print("Presecna ucestanost pojacanja:",round(w1,2))
```

```python
#fazna margina
PM=180+phase[mag[:]>=0][-1]
print("Fazna margina:",round(PM,2))
```

```python
# presecna ucestanost faze je beskonacna 
w_pi=w[phase[:]>=-180][-1]
print("Presecna ucestanost faze:",round(w_pi,2))
```

```python
# amplitudska margina je beskonacna
d=1/10**((mag[phase[:]>=-180][-1])/20)
print("Amplitudska margina:",round(d,2))
```

```python
#c) Teorijski je sistem stabilan za bilo koje K
```

```python
#d) 
mag_PM=mag[phase[:]>=-180+30][-1]
K=1/(10**(mag_PM/20))
print("Pojacanje za koje je PM 30 je",round(K,1))
```

```python
#e)
W1=signal.lti([], [0, -200], K*2000)
ww, mag1, phase1 = signal.bode(W1,w=w_log)
plt.figure()
plt.grid()
plt.semilogx(ww, mag1)    # Bode amplitudska karakteristika
plt.figure()
plt.grid()
plt.semilogx(ww, phase1)  # Bode fazna karakteristika
plt.show()
```
