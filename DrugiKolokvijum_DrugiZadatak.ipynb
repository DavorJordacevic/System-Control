```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from scipy import signal
```

```python
W= signal.lti([],[-5,-3,-3],1)
w_log= np.logspace(-2,4,num=1000)
w, mag, phase= signal.bode(W,w=w_log)
plt.figure()
plt.grid()
plt.semilogx(w,mag)
plt.figure()
plt.grid()
plt.semilogx(w,phase)
plt.show()
```

```python
ww,H= signal.freqresp(W)
plt.figure()
plt.grid()
plt.plot(H.real, H.imag, "b")
plt.plot(H.real, -H.imag, "r")
plt.show()
```

```python
# C(s)= K
K=5
W= signal.lti([],[-5,-3,-3],1*K)
w_log= np.logspace(-2,4,num=1000)
w, mag, phase= signal.bode(W,w=w_log)
plt.figure()
plt.grid()
plt.semilogx(w,mag)
plt.figure()
plt.grid()
plt.semilogx(w,phase)
plt.show()
ww,H= signal.freqresp(W)
plt.figure()
plt.grid()
plt.plot(H.real, H.imag, "b")
plt.plot(H.real, -H.imag, "r")
plt.scatter(-1/K,0,marker='o',color='g')
plt.show()
```
